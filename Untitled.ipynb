```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import numpy as np
import cv2
import os
from sklearn.cluster import KMeans







        
# Fonction pour ouvrir une image
def open_image():
    global image, photo, original_image, original_photo

    # Obtenir le nom de l'image à partir de l'Entry
    selected_image_name = entry_image_name.get()

    if selected_image_name:
        # Ajouter le chemin complet du dossier où se trouvent les images si nécessaire
        # Remplacez "chemin/vers/dossier/contenant/les/images/" par le chemin de votre dossier
        selected_image_path = f"{selected_image_name}"

        # Vérifier si le fichier image existe
        if os.path.exists(selected_image_path):
            image = Image.open(selected_image_path)
            image.thumbnail((800, 300))
            photo = ImageTk.PhotoImage(image)

            # Assurez-vous que la référence à l'image est correctement gérée ici
            canvas.image = photo
            canvas.create_image(0, 0, anchor=tk.NW, image=photo)

            original_image = image.copy()
            original_photo = ImageTk.PhotoImage(original_image)

            # Assurez-vous que la référence à l'image originale est correctement gérée ici
            original_canvas.image = original_photo
            original_canvas.create_image(0, 0, anchor=tk.NW, image=original_photo)
        else:
            print(f"Le fichier {selected_image_path} n'existe pas.")
    else:
        print("Veuillez entrer le nom de l'image.")


# Fonction pour afficher l'image originale
def display_original():
    global original_image, original_photo
    if original_image and original_photo:
        original_canvas.delete("all")
        original_canvas.image = original_photo
        original_canvas.create_image(0, 0, anchor=tk.NW, image=original_photo)

def morphologie_fermeture():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        kernel = np.ones((5, 5), np.uint8)  # Vous pouvez ajuster la taille du noyau
        fermeture = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)
        fermeture_pil = Image.fromarray(fermeture)
        photo = ImageTk.PhotoImage(fermeture_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        
def contour_sobel():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)  # Sobel X
        sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)  # Sobel Y
        sobel = cv2.sqrt(cv2.addWeighted(cv2.pow(sobel_x, 2.0), 1.0, cv2.pow(sobel_y, 2.0), 1.0, 0.0))
        sobel_pil = Image.fromarray(sobel)
        photo = ImageTk.PhotoImage(sobel_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        
def contour_robinson():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        robinson_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
        robinson_y = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

        robinson_x_filtered = cv2.filter2D(gray, cv2.CV_64F, robinson_x)
        robinson_y_filtered = cv2.filter2D(gray, cv2.CV_64F, robinson_y)

        robinson = cv2.sqrt(cv2.addWeighted(cv2.pow(np.abs(robinson_x_filtered), 2.0), 1.0, cv2.pow(np.abs(robinson_y_filtered), 2.0), 1.0, 0.0))
        robinson_pil = Image.fromarray(robinson)
        photo = ImageTk.PhotoImage(robinson_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)



def contour_prewit():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        prewit_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)  # Prewit X
        prewit_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)  # Prewit Y
        prewit = cv2.sqrt(cv2.addWeighted(cv2.pow(prewit_x, 2.0), 1.0, cv2.pow(prewit_y, 2.0), 1.0, 0.0))
        prewit_pil = Image.fromarray(prewit)
        photo = ImageTk.PhotoImage(prewit_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)



def morphologie_ouverture():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        kernel = np.ones((5, 5), np.uint8)  # Vous pouvez ajuster la taille du noyau
        ouverture = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
        ouverture_pil = Image.fromarray(ouverture)
        photo = ImageTk.PhotoImage(ouverture_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)



def contour_laplacian():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        laplacian = cv2.Laplacian(gray, cv2.CV_64F)
        laplacian_pil = Image.fromarray(laplacian)
        photo = ImageTk.PhotoImage(laplacian_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
def apply_otsu():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        _, binarized = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        binarized_pil = Image.fromarray(binarized)
        photo = ImageTk.PhotoImage(binarized_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        
def apply_seuillage_fixe():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        _, binarized = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY)  # Seuillage fixe à 128 
        binarized_pil = Image.fromarray(binarized)
        photo = ImageTk.PhotoImage(binarized_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
def apply_filtrage_moyen():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        filtered = cv2.blur(image_cv, (5, 5))  # Filtrage moyen avec un noyau 5x5
        filtered_pil = Image.fromarray(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
        photo = ImageTk.PhotoImage(filtered_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
def apply_filtrage_median():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        filtered = cv2.medianBlur(image_cv, 5)  # Filtrage médian avec un noyau de taille 5
        filtered_pil = Image.fromarray(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
        photo = ImageTk.PhotoImage(filtered_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)


def morphologie_erosion():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        kernel = np.ones((5, 5), np.uint8)  # Vous pouvez ajuster la taille du noyau
        erosion = cv2.erode(gray, kernel, iterations=1)
        erosion_pil = Image.fromarray(erosion)
        photo = ImageTk.PhotoImage(erosion_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
def morphologie_dilatation():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        kernel = np.ones((5, 5), np.uint8)  # Vous pouvez ajuster la taille du noyau
        dilatation = cv2.dilate(gray, kernel, iterations=1)
        dilatation_pil = Image.fromarray(dilatation)
        photo = ImageTk.PhotoImage(dilatation_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        
def morphologie_contour():
    global image, photo
    if image:
        image_cv = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
        gray = cv2.cvtColor(image_cv, cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        contour_image = np.zeros_like(gray)
        cv2.drawContours(contour_image, contours, -1, (255, 255, 255), 1)
        contour_image_pil = Image.fromarray(contour_image)
        photo = ImageTk.PhotoImage(contour_image_pil)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)

def region_growth_and_display():
    global image, photo, canvas

    # Récupérer les coordonnées du point de départ depuis l'utilisateur
    seed_x = int(entry_seed_x.get())
    seed_y = int(entry_seed_y.get())
    seed_point = (seed_x, seed_y)

    if image:
        # Convertir l'image en tableau numpy
        image_array = np.array(image)

        visited = np.zeros_like(image_array[:, :, 0], dtype=np.uint8)
        stack = [seed_point]
        seed_color = image_array[seed_point[1], seed_point[0]]

        while stack:
            current_point = stack.pop()
            if visited[current_point[1], current_point[0]] == 0 and np.all(
                    image_array[current_point[1], current_point[0]] == seed_color):
                visited[current_point[1], current_point[0]] = 255
                for i in range(-1, 2):
                    for j in range(-1, 2):
                        new_point = (current_point[0] + i, current_point[1] + j)
                        if 0 <= new_point[0] < image_array.shape[1] and 0 <= new_point[1] < image_array.shape[0]:
                            stack.append(new_point)

        segmented_image = Image.fromarray(visited)

        # Afficher l'image segmentée sur le canvas
        photo = ImageTk.PhotoImage(segmented_image)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)




def partition_region():
    global image, photo, canvas

    if image:
        # Convertir l'image en tableau numpy
        image_array = np.array(image)
        # Remodeler l'image pour avoir une liste de pixels
        pixels = image_array.reshape((-1, 3))
        # Déterminer le nombre de clusters (régions)
        num_clusters = 5  # Vous pouvez ajuster ce nombre en fonction de vos besoins
        # Appliquer K-means pour partitionner l'image
        kmeans = KMeans(n_clusters=num_clusters, random_state=0)
        labels = kmeans.fit_predict(pixels)
        # Remodeler les étiquettes pour obtenir la même forme que l'image
        segmented_image_array = labels.reshape(image_array.shape[:2])
        # Créer une image à partir de l'array segmenté
        segmented_image = Image.fromarray((segmented_image_array * (255 / (num_clusters - 1))).astype('uint8'))
        # Afficher l'image segmentée sur le canvas
        photo = ImageTk.PhotoImage(segmented_image)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
        



def fusion_segmentation():
    global image, photo

    if image:
        # Convertir l'image en tableau numpy
        image_array = np.array(image)

        # Déterminer le nombre de clusters (régions)
        num_clusters = 2  # Vous pouvez ajuster ce nombre en fonction de vos besoins

        # Appliquer K-means pour partitionner l'image
        kmeans = KMeans(n_clusters=num_clusters, random_state=0, n_init=10)  # Explicitly set n_init
        labels = kmeans.fit_predict(image_array.reshape((-1, 3)))

        # Remodeler les étiquettes pour obtenir la même forme que l'image
        segmented_image_array = labels.reshape(image_array.shape[:2])

        # Créer une copie de l'image pour éviter de modifier l'original
        segmented_image = np.copy(image_array)

        # Appliquer la segmentation par fusion basée sur la couleur
        for label in range(num_clusters):
            # Créer un masque pour le cluster courant
            mask = (segmented_image_array == label)

            # Appliquer la segmentation par fusion sur la copie de l'image
            mean_color = np.mean(image_array[mask], axis=(0, 1))  # Calculate mean color across all channels
            segmented_image[mask] = mean_color.astype('uint8')

        

        # Créer une image à partir de l'array segmenté après la fusion
        segmented_image = Image.fromarray(segmented_image)

        # Afficher l'image segmentée sur le canvas
        photo = ImageTk.PhotoImage(segmented_image)
        canvas.image = photo
        canvas.create_image(0, 0, anchor=tk.NW, image=photo)
    else:
        print("Aucune image chargée pour la segmentation fusion.")





def create_section(title_text, x, y):
    global entry_seed_x, entry_seed_y
    section_frame = tk.Frame(root, padx=10, pady=10, bg='#f0f0f0', bd=2, relief=tk.RAISED)
    section_frame.place(x=x, y=y)

    title_label = tk.Label(section_frame, text=title_text, font=("Helvetica", 14, "bold"), bg='#f0f0f0')
    title_label.pack()

    buttons_frame = tk.Frame(section_frame, bg='#f0f0f0')
    buttons_frame.pack()

    button1 = None
    button2 = None
    button3 = None
    button4 = None
    button5 = None
    

    if title_text == "Binarisation":
        button1 = tk.Button(buttons_frame, text="Otsu", bg='#add8e6', fg='black', command=apply_otsu)
        button2 = tk.Button(buttons_frame, text="Seuillage Fixe", bg='#add8e6', fg='black', command=apply_seuillage_fixe)
    elif title_text == "Filtrage":
        button1 = tk.Button(section_frame, text="Moyen", bg='#add8e6', fg='black', command=apply_filtrage_moyen)
        button2 = tk.Button(section_frame, text="Median", bg='#add8e6', fg='black', command=apply_filtrage_median)
    elif title_text == "Extraction des contours":
        button1 = tk.Button(section_frame, text="Sobel", bg='#add8e6', fg='black', command=contour_sobel)
        button2 = tk.Button(section_frame, text="Robinson", bg='#add8e6', fg='black', command=contour_robinson)
        button3 = tk.Button(section_frame, text="Prewit", bg='#add8e6', fg='black', command=contour_prewit)
        button4 = tk.Button(section_frame, text="Laplacian", bg='#add8e6', fg='black', command=contour_laplacian)
    elif title_text == "Morphologie Mathematique":
        button1 = tk.Button(section_frame, text="Erosion", bg='#add8e6', fg='black', command=morphologie_erosion)
        button2 = tk.Button(section_frame, text="Dilatation", bg='#add8e6', fg='black', command=morphologie_dilatation)
        button3 = tk.Button(section_frame, text="Ferméture", bg='#add8e6', fg='black', command=morphologie_fermeture)
        button4 = tk.Button(section_frame, text="Ouverture", bg='#add8e6', fg='black', command=morphologie_ouverture)
        button5 = tk.Button(section_frame, text="Contour", bg='#add8e6', fg='black', command=morphologie_contour)
    elif title_text=="Croissance de Régions":
        button1 = tk.Button(section_frame, text="Croissance  Régions", bg='#add8e6', fg='black', command=region_growth_and_display)
        label_seed = tk.Label(section_frame, text="Coordonnées du point de départ :", font=("Helvetica", 12), bg='#f0f0f0')
        label_seed.pack()
        label_seed_x = tk.Label(section_frame, text="X :", font=("Helvetica", 10), bg='#f0f0f0')
        label_seed_x.pack()
        entry_seed_x = tk.Entry(section_frame, font=("Helvetica", 10), width=5)
        entry_seed_x.insert(0, "10")
        entry_seed_x.pack()
        label_seed_y = tk.Label(section_frame, text="Y :", font=("Helvetica", 10), bg='#f0f0f0')
        label_seed_y.pack()
        entry_seed_y = tk.Entry(section_frame, font=("Helvetica", 10), width=5)
        entry_seed_y.insert(0, "50")
        entry_seed_y.pack()

    else:
        button1 = tk.Button(section_frame, text="Partition de Régions", bg='#add8e6', fg='black', command=partition_region)
        button2 = tk.Button(section_frame, text="Segmentation Fusion", bg='#add8e6', fg='black', command=fusion_segmentation)

       
    if button1:
        button1.pack(side=tk.LEFT, padx=5)
    if button2:
        button2.pack(side=tk.LEFT, padx=5)
    if button3:
        button3.pack(side=tk.LEFT, padx=5)
    if button4:
        button4.pack(side=tk.LEFT, padx=5)
    if button5:
        button5.pack(side=tk.LEFT, padx=5)
```
